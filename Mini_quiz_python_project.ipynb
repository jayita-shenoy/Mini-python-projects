```python



questions = ( "Which planet in our solar system is known as the red planet? ",
              " Which river is known as the longest in the world?",
              "What food do Koalas usually eat?",
              "What is the name of the science that studies the past of life on earth?",
              "What element is used in the production of steel?")
options = (("A. Mercury","B. Venus","C. Earth","D. Mars "),
           ("A. Nile","B. Danube","C. Yangtze","D. Amazon"),
           ("A. Bamboos","B. Eucalyptus","C. Aloe Vera ","D. Bananas"),
           ("A. Palaeontology","B. Astronomy ","C. Anthropology","D. Geology"),
           ("A. Copper","B. Tin","C. Iron","D. Silver"))
answers = ("D", "A", "B", "A", "C")
guesses = []


print("Welcome! LET'S BEGIN!")
print("")
print("please take note that only your first correct guess will be taken into account in your final score ")
print("")
print("")
question_num = 0

for question in questions:

  print("---------------------------------------")
  print(question)
  for option in options[question_num]:
    print(option)

  guess_1 = input("Choose option A, B, C, D").upper()

  guesses.append(guess_1)

  if guess_1 == answers[question_num]:

     print("your answer is CORRECT!")

     question_num += 1
  else:

    while not guess_1 == answers[question_num]:
      print("Your answer is WRONG! TRY AGAIN")

      guess_2 = input("Choose option A, B, C, D").upper()
      if guess_2 == answers[question_num]:
       print("your answer is CORRECT!")

       break

    question_num += 1



  print("---------------------------------------")


answers_list = list(answers)





correct_guesses = sum([1 for i in range(len(answers)) if guesses[i] == answers[i]])


print(f" YOUR FINAL SCORE IS: {correct_guesses} / 5!")













```
